```python
import matplotlib.pyplot as plt
  
year = [1972, 1982, 1992, 2002, 2012]
e_india = [100.6, 158.61, 305.54, 
           394.96, 724.79]
  
e_bangladesh = [10.5, 25.21, 58.65,
                119.27, 274.87]
plt.xlabel('years')
plt.ylabel('power consumption per year')
plt.title('Electricity Consumption Data')

plt.plot(year,e_india,color='r',marker='o',markersize=6,label='india');
plt.plot(year,e_bangladesh,color='g',linestyle='dashed',linewidth='3',label='bangladesh')
plt.legend()
plt.show()
```
